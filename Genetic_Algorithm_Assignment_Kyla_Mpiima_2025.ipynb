## Discussion and justification on the approaches taken for the above

the key aspects of the individual class is its crossover method, which combines genetic material from two parent individuals to produce offspring. In this implementation, a fixed crossover point is used to splice the chromosomes of the two parents. Specifically, the method joins the beginning segment of one parent's chromosome with the ending segment of the other parent. This approach is particularly important for audio waveforms, as the sequential nature of audio data means that maintaining the order of samples is crucial for preserving sound quality.

Traditional crossover methods, which might randomly swap segments of chromosomes, are not suitable for audio data because they can disrupt the  coherence necessary for the waveform to sound natural. By using a fixed crossover point, the method ensures that the resulting offspring inherit a continuous segment of audio data from each parent, thereby maintaining the integrity of the waveform. This is essential for generating audio that is not only similar to the target but also coherent and listenable


```python
class individual:
    def __init__(self, prob):
      self.chromosome = np.random.uniform(prob.min_value, prob.max_value,prob.number_of_genes)
      self.cost = prob.cost_function(self.chromosome)
    def mutate(self, rate_of_gene_mutation, range_of_gene_mutation):
        for index in range(len(self.chromosome)):
            if np.random.uniform() < rate_of_gene_mutation:
                self.chromosome[index] += np.random.randint(-100, 101) * range_of_gene_mutation

    def crossover(self, parent2, explore_crossover):
      crossoverPoint = 1250; #random midpoint
      child1 = deepcopy(self)
      child2 = deepcopy(parent2)
      child1.chromosome = np.append(self.chromosome[:crossoverPoint] , parent2.chromosome[crossoverPoint:])
      child2.chromosome = np.append( parent2.chromosome[:crossoverPoint] , self.chromosome[crossoverPoint:])

      return child1, child2
```

```python

class parameters:
  def __init__(self):
    self.population  = 1000
    self.explore_crossover_range = 0.2
    self.gene_mutate_range = 1
    self.gene_mutate_rate = 0.2
    self.birth_rate_per_generation = 1
    self.max_number_of_generations = 1000
```

```python
def choose_parents(population):
  index1 = np.random.randint(0,len(population))
  index2 = np.random.randint(0,len(population))
  if index1 == index2:
    return choose_parents(population)
  return index1, index2
```

```python
from copy import deepcopy

def run_genetic(prob,params):
  number_in_population = params.population
  explore_crossover = params.explore_crossover_range
  gene_mutate_rate = params.gene_mutate_rate
  max_number_of_generations = params.max_number_of_generations
  gene_mutate_range = params.gene_mutate_range
  cost_function = prob.cost_function
  number_of_children_per_generation = number_in_population * params.birth_rate_per_generation
  acceptable_cost = prob.acceptable_cost

  population = []
  best_solution = individual(prob)
  best_solution.cost = np.inf
  for i in range(number_in_population):
    new_individual = individual(prob)
    if new_individual.cost:
      best_solution = deepcopy(new_individual)
    population.append(individual(prob))

  for i in range(max_number_of_generations):
    children = []

    while len(children) < number_of_children_per_generation:
      parent1_index, parent2_index = choose_parents(population)

      parent1 = population[parent1_index]
      parent2 = population[parent2_index]

      child1, child2 = parent1.crossover(parent2,explore_crossover)
      child1.mutate(gene_mutate_rate, gene_mutate_range)
      child1.cost = cost_function(child1.chromosome)
      child2.mutate(gene_mutate_rate, gene_mutate_range)
      child2.cost = cost_function(child2.chromosome)

      children.append(child1)
      children.append(child2)

    population += children

    population = sorted(population, key=lambda x: x.cost)

    population = population[:number_in_population]

    if population[0].cost < best_solution.cost:
      best_solution = deepcopy(population[0])
      print(best_solution.cost)
    if best_solution.cost < acceptable_cost:
      break

  return population, best_solution
```

The saveWav function is crucial for exporting the best audio solution generated by a genetic algorithm into a WAV file format, enabling playback and analysis. It calculates the total number of samples based on the sample rate and duration, initializes an audio data array filled with the best solution's cost, and clips the data to ensure it remains within the valid range for 16-bit PCM audio. The function then converts the data to the appropriate type and writes it to a WAV file using the write function. Finally, it provides user feedback to confirm successful file creation. Overall, this function ensures that the generated audio is preserved in a standard format while maintaining quality, making it essential for evaluating the algorithm's performance.

```python
def saveWav(best_solution,sample_rate, duration, output_filename="modified.wav"):
    num_samples = int(sample_rate * duration)
    audio_data = np.full(num_samples, best_solution.cost, dtype=np.float64)


    audio_data = np.clip(audio_data, -32768, 32767)
    audio_data = audio_data.astype(np.int16)

    write(output_filename, sample_rate, audio_data)

    print(f"WAV file '{output_filename}' created ")
```
